## The isin() method checks if the Dataframe contains the specified values.

```python
def Item_Outlet_sales(x):
    if x<=1500:
        return 'low'
    if x>1500 and x<6000:
        return 'medium'
    else:
        return 'high'
```

```python
file['sales_category']=file['Item_Outlet_Sales'].apply(Item_Outlet_sales)
file.head()
```

## The apply() method allows user to apply a function along one of the axis of the Dataframe,also it returns new Dataframe object after applying the function to its elements.

```python
file.drop('sales_category',axis=1,inplace=True)
```
